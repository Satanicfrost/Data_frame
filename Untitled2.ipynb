```python
df= pd.read_excel(r'C:\Users\Leonardo\Documents\zno.xlsx', sheet_name = None, ignore_index=True)
df = pd.concat(df.values())

print(df)
```

```python
df.columns = df.columns.str.lower()
df.rename(columns =
          {
              'parâmetro de rede':'a',
              'unnamed: 2': 'c',
              'tamanho de cristalito (nm) ':'tamanho',
              'volume da célula unitária (a³) ':'volume'
          },inplace=True)
df
```

```python
df = df.dropna()
```

```python
df['temperatura'] = df['amostra'].apply(lambda x: x[-5:-2])
df['porcentagem'] = df['amostra'].apply(lambda x: x.split()[1][:-1].replace(',','.'))
df
```

```python
df =df[['amostra', 'porcentagem', 'temperatura', 'a', 'c', 'c/a', 'tamanho', 'volume']]
```

```python
df['volume'] = df['volume'].apply(lambda x: x/1000)
df =df.round(2)                            
df['a'] = df['a'].apply(lambda x: round(x, 2))
df['c'] = df['c'].apply(lambda x: round(x, 2))
df
```

```python
df.drop('amostra',1, inplace=True)
```

```python
df.reset_index(drop=True)
```

```python
df = df.apply(pd.to_numeric, errors='coerce')
df['a'] = df['a'].apply(lambda x: round(x, 2))
df['c'] = df['c'].apply(lambda x: round(x, 2))
df['porcentagem'] = df['porcentagem'].apply(lambda x: round(x, 2))
df['temperatura'] = df['temperatura'].apply(lambda x: round(x, 1))
df['tamanho'] = df['tamanho'].apply(lambda x: round(x, 2))
df['volume'] = df['volume'].apply(lambda x: round(x, 2))

df
```
